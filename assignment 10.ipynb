```python
import numpy as np


a = 0.1
b = 0.4
c = 0.7
d = 0.9

def explore_arm():
    arm1 = np.random.uniform(a,c,1)
    arm2 = np.random.uniform(b,d,1)
    if arm1>arm2:
        return(['arm1',arm1,arm2])
    else:
        return(['arm2',arm1,arm2])

    
def C():
    print('*********** Starting a new game ***********')
    arm = explore_arm()
    reward = []
    if arm[0]=='arm1' :
        print('exploiting arm1')
        for i in range(20):
            reward.append(np.random.uniform(a,c,1))
    else:
        print('exploiting arm2')
        for i in range(20):
            reward.append(np.random.uniform(b,d,1))
    reward.append(arm[1])
    reward.append(arm[2])
    print('Total reward for this game is: ',sum(reward))
    print('*********** End *********** \n\n')
    return(sum(reward))
```

```python
for i in range(10):
    C()
```
